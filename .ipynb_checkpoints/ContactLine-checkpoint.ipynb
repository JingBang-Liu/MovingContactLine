## First decide position then average

Suppose the position of contact line $X$ is a function of $y$ and time $t$, that is $X(y,t)$. Here we try to extract $X$ from the density field. Information of density field is stored in a $T\times NBX\times NBY$ matrix $B$. 

### Least square fitting using tanh

```python
# correlation of contactline in miniscus
def tanh(x, a, b, c, d):
    return a*(np.exp(b*(x - c))-1)/(np.exp(b*(x - c))+1) + d


#### Maybe only 3 parameters

x1 = np.zeros(NBX)
x2 = np.zeros(NBX)
y1 = np.zeros(NBY)
for i in range(NBX):
    x1[i] = XD1[0]+dx/2+i*dx
for i in range(NBX):
    x2[i] = XD2[0]+dx/2+i*dx
for i in range(NBY):
    y1[i] = YD[0]+dy/2+i*dy
```

```python
# fitting using least square
t = 2
ny = 21
ydata = left_bins[t,:29,ny,0] + left_bins[t,:29,ny,1]
#ydata = right_bins[t,:,ny,0] + right_bins[t,:,ny,1]

para_bound_left = {}
para_bound_left[0] = [-np.inf, 0.01, 65, 1]
para_bound_left[1] = [+np.inf, +np.inf, 75, +np.inf]
para_bound_right = {}
para_bound_right[0] = [-np.inf, -np.inf, 135, 1]
para_bound_right[1] = [+np.inf, -0.01, 145, +np.inf]

para_bound = para_bound_left
#para_bound = para_bound_right
x = x1[:29]
#x = x2
popt, pcov = curve_fit(tanh, x, ydata, bounds=(para_bound[0],para_bound[1]))
#popt, pcov = curve_fit(tanh, x1[:29], ydata)
```

```python
plt.plot(x, tanh(x, *popt),
         'r-', label='fit: a=%5.3f, b=%5.3f, c=%5.3f, d=%5.3f' % tuple(popt))
plt.plot(x, ydata, '-b', label='data')
plt.xlabel('x')
plt.ylabel('dens')
plt.legend()
plt.title('Fitting illustration')
plt.show()
perr = np.sqrt(np.diag(pcov))
perr
```

```python
# function that determins the position of contact line by fitting to tanh and half density
def con_pos(dens,side):
    if side == 'left':
        x = x1[:29]
        para_bound = para_bound_left
        dens = dens[:29]
    else:
        x = x2
        para_bound = para_bound_right
    popt, pcov = curve_fit(tanh, x, dens, bounds=(para_bound[0],para_bound[1]))
    return popt
```

```python
CL_left = np.zeros((timemarks,NBY))
for i in range(timemarks):
    for j in range(NBY):
        dens = left_bins[i,:,j,0] + left_bins[i,:,j,1]
        popt, pcov = curve_fit(tanh, x1[:29], dens[:29], bounds=(para_bound_left[0],para_bound_left[1]))
        CL_left[i,j] = popt[2]
        
CL_right = np.zeros((timemarks,NBY))
for i in range(timemarks):
    for j in range(NBY):
        dens = right_bins[i,:,j,0] + right_bins[i,:,j,1]
        popt, pcov = curve_fit(tanh, x2, dens, bounds=(para_bound_right[0],para_bound_right[1]))
        CL_left[i,j] = popt[2]
```

### Find correlation between moving contact lines

In this section, we would like to explore the correlation between different contact lines as well as their autocorrelation functions. For autocorrelation, it makes more sense to work on displacement of contact lines to their mean postion.

```python
# calculation displacement
CL_dis_left_bot = CL_left_bot_avg - mu_left_bot
CL_dis_left_top = CL_left_top_avg - mu_left_top
CL_dis_right_bot = CL_right_bot_avg - mu_right_bot
CL_dis_right_top = CL_right_top_avg - mu_right_top
```

```python
plt.plot(CL_dis_right_bot[::50],'-r',label='bottom')
plt.plot(CL_dis_right_top[::50],'-b',label='top')
plt.xlabel('t')
plt.ylabel('Displacement of CL')
plt.title('Displacement of right CL over time')
plt.legend()
plt.show()
```

```python
# Time average function by Blake
def time_avg_1(x):
    T = np.size(x)
    #a = np.ones(T)
    #full_mean = np.linspace(T,1,T)
    #result = np.correlate(a,x,mode='full')
    #return np.divide(result[result.size // 2:],full_mean)
    result = np.zeros(T)
    for i in range(T):
        result[i] = np.mean(x[i::])
    return result
# define autocorrelation function
def autocorr(x):
    T = np.size(x)
    full_mean = np.linspace(T,1,T)
    result = np.correlate(x,x,mode='full')
    return np.divide(result[result.size // 2:],full_mean)
# define correlation function
def corr(x,y):
    T = np.size(x)
    full_mean = np.linspace(T,1,T)
    result = np.correlate(x,y,mode='full')
    return np.divide(result[result.size // 2:],full_mean)
```

```python
# test for autocorrelation function
plt.plot(corr(CL_dis_right_bot[::50],CL_dis_right_top[::50]))
```

```python
plt.plot(time_avg_1(np.abs(CL_dis_right_top[::50])))
plt.xlabel('time step')
plt.ylabel('average')
plt.title('Blakes average')
plt.show()
T_tol = 5*2000000*1e-6
print('Total run time is ', T_tol, 'ns')
```
